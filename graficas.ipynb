```python
import numpy as np
import matplotlib.pyplot as plt 

from matplotlib.pyplot import plot,show,xlim,ylim,xticks, legend, figure,subplot
from numpy import linspace,pi,sin,cos,tan,log
```

```python

# programa para graficar una recta ,con base en la encuacion ,
#m es la pendiente de la recta
#b es el punto de corte con el eje y


print("----------------------------")
print("-----funciones:----")





print("\nDame la opcion que deseas realizar:\n")
print("-----------------------------")
print("--1.FUNCION LNEAL------------")
print("--2.FUNCION CUADRATICA-------")
print("--3.FUNCION CUBICA-----------")
print("--4.FUNCION EXPONENCIAL------")
print("--5.FUNCION LOGARITMICA------")
print("--6.FUNCION TRIGONOMETRICAS--")


opcion=int(input("\nDame la opcion:"))

if (opcion==1):
  N=100

  #funcion para calcular los valores de y
  def funcion_lineal (m,b,x):
    return m*x + b

  m=int(input("Digite el valor de la pendiente: "))
  b=int(input("Digite el valor del punto de corte con el eje y : "))
 
  #processing- grafica
  x=np.linspace(-10,10,num=N)
  y=funcion_lineal(m,b,x)

  plt.plot(x,y , color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("funcion lineal : y=mx+b")
  plt.grid()
  plt.axhline(y=0 ,color="b")
  plt.show()

elif (opcion ==2):
# funcion cuadratica
  N=100

  def funcion_cuadratica (a,b,x,c):
    return a*(x**2)+(b*x)+c

  a=int(input("Digite el valor de a: "))
  b=int(input("Digite el valor de b : "))
  c=int(input("Digite el valor de c: "))
 
  #processing- grafica
  x=np.linspace(-10,10,num=N)
  y=funcion_cuadratica(a,b,x,c)

  plt.plot(x,y , color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("funcion cuadratica y = Ax**2 +b*x +c ")
  plt.grid()
  plt.axhline(y=0 ,color="b")
  plt.axvline(x=0, color="b")
  plt.show()


elif (opcion ==3):
  # funcion cubica
  N=100

  def funcion_cubica (a,x,b,c,d):
    return a*(x**3)+b*(x**2)+(c*x)+d

  a=int(input("Digite el valor de a: "))
  b=int(input("Digite el valor de b : "))
  c=int(input("Digite el valor de c: "))
  d=int(input("Digite el valor de d: "))
    
  #processing- grafica
  x=np.linspace(-10,10,num=N)
  y=funcion_cubica(a,x,b,c,d)

  plt.plot(x,y , color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("funcion cuadratica y = ax**3 +bx**2+cx+d ")
  plt.grid()
  plt.axhline(y=0 ,color="b")
  plt.axvline(x=0, color="b")
  plt.show()


elif(opcion==4):
  N=100
  
  def funcion_exponencial (a,x):
   return a**x 
   
  a=int(input("Digite el valor de a: "))
  
  #processing- grafica
  x=np.linspace(-10,10,num=N)
  y=funcion_exponencial(a,x,)

  plt.plot(x,y , color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("funcion exponencial Y=a**x ")
  plt.grid()
  plt.axhline(y=0 ,color="b")
  plt.axvline(x=0, color="b")
  plt.show()

elif (opcion==5):
  N=100
  
  def funcion_logaritmica (x):
   return np.log(x)
   
  x=int(input("Digite el valor de x: "))
  
  #processing- grafica
  x=np.linspace(-10,10,num=N)
  y=funcion_logaritmica(x)

  plt.plot(x,y , color="r")
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title("funcion exponencial Y= log(x)")
  plt.grid()
  plt.axhline(y=0 ,color="b")
  plt.axvline(x=0, color="b")
  plt.show()


elif (opcion==6):
  N=100
  print("--1.Seno--")
  print("--2.Coseno--")
  print(("--3.Tangente"))
  opcion=int(input("Digite la funcion que deseas realizar: "))
  
  if (opcion==1):
    N=100
    def funcion_trigonometricas (sin, x):
      return sin(x)
   
  
  
    #processing- grafica
    x=linspace(-pi,pi,num=N)
    y=sin(x)
    legend(loc="upper left")
    plot(x,y, label="seno")
    ylim(-1,1)
    plt.axhline(y=0 ,color="b")
    plt.axvline(x=0, color="b")
    xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
    show()

  elif(opcion==2):
    N=100
   
    def funcion_trigonometricas (cos, x):
      return cos(x)
   
  
    #processing- grafica
    x=linspace(-pi,2*pi,num=N)
    y=cos(x)
    legend(loc="upper left")
    plot(x,y, label="coseno")
    ylim(-1,1)
    plt.axhline(y=0 ,color="b")
    plt.axvline(x=0, color="b")
    xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
    show()
  
  elif(opcion==3):
    N=100

    def funcion_trigonometricas (tan, x):
      return tan(x)
   
  
  
    #processing- grafica
    x=linspace(-pi,pi,num=N)
    y=tan(x)
    legend(loc="upper left")
    plot(x,y, label="Tangente")
    ylim(-10,10)
    plt.axhline(y=0 ,color="b")
    plt.axvline(x=0, color="b")
    xticks([-pi,-pi/2,0,pi/2,pi],[r'$-\pi$',r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$'])
    show()

```
